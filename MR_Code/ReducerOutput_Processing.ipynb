```python
import pandas as pd
```

### WordCount Output Processing

```python
df1 = pd.read_csv('../Code/Word_Count/part-00000', delimiter="\t", names= ["word", "count"])
df2 = pd.read_csv('../Code/Word_Count/part-00001', delimiter="\t", names= ["word", "count"])
df3 = pd.read_csv('../Code/Word_Count/part-00002', delimiter="\t", names= ["word", "count"])
df4 = pd.read_csv('../Code/Word_Count/part-00003', delimiter="\t", names= ["word", "count"])
df5 = pd.read_csv('../Code/Word_Count/part-00004', delimiter="\t", names= ["word", "count"])
df6 = pd.read_csv('../Code/Word_Count/part-00005', delimiter="\t", names= ["word", "count"])
df7 = pd.read_csv('../Code/Word_Count/part-00006', delimiter="\t", names= ["word", "count"])

final_df = pd.concat([df1, df2, df3, df4, df5, df6, df7])
final_df = final_df.sort_values(by="count", ascending=False)
final_df.to_csv('../Code/Word_Count/NYT_wordcounts.csv', index = False)
```

### WordCooccurence Output Processing

```python
df_1 = pd.read_csv('../Code/Word_Cooccurence/part-00000', delimiter="\t", names= ["bigram", "count"])
df_2 = pd.read_csv('../Code/Word_Cooccurence/part-00001', delimiter="\t", names= ["bigram", "count"])
df_3 = pd.read_csv('../Code/Word_Cooccurence/part-00002', delimiter="\t", names= ["bigram", "count"])

final_df2 = pd.concat([df_1, df_2, df_3])
final_df2 = final_df2.sort_values(by="count", ascending=False)
final_df2.to_csv('../Code/Word_Cooccurence/NYT_Cooccurence_counts.csv', index = False)
```
